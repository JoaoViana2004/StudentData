```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

dados_bruto = pd.read_csv("/content/students_mental_health_survey.csv")
dados_bruto.head()
```

##**Limpando DataSet**


```python
dados = dados_bruto.dropna()
dados.info()
```

```python
stress = "Stress_Level"
depre = "Depression_Score"
ansiedade = "Anxiety_Score"
social = "Social_Support"
sono = "Sleep_Quality"
diet = "Diet_Quality"
relac = "Relationship_Status"
subs = "Substance_Use"
creditos = "Semester_Credit_Load"
curso ="Course"
casa = "Residence_Type"
aconselhamento = "Counseling_Service_Use"
atvdd_fisica = 'Physical_Activity'
```

```python
cont = 2

for x in dados[sono].unique():
  dados[sono] = dados[sono].replace(x, cont)
  cont -= 1

dados[sono].unique()
```

```python
cont = 0

for x in dados[aconselhamento].unique():
  dados[aconselhamento] = dados[aconselhamento].replace(x, cont)
  cont += 1

dados[aconselhamento].unique()
```

```python
cont = 0

for x in ['Low','Moderate','High']:
  dados[atvdd_fisica] = dados[atvdd_fisica].replace(x, cont)
  cont += 1

dados[atvdd_fisica].unique()
```

3- Qual é o impacto das atividades físicas na depressão dos estudantes ??

```python
for c in dados[curso].unique():
  print(f'Para o curso {c}:')
  print(f'Media dos Niveis de Depressao de alunos que não Fazem exercicios fisicos: {dados[(dados[curso] == c) & (dados[atvdd_fisica] == 0)][depre].mean():.3f}')
  print(f'Media dos Niveis de Depressao de alunos que fazem alguns exercicios fisicos: {dados[(dados[curso] == c) & (dados[atvdd_fisica] == 1)][depre].mean():.3f}')
  print(f'Media dos Niveis de Depressao de alunos que fazem muitos exercicios fisicos: {dados[(dados[curso] == c) & (dados[atvdd_fisica] == 2)][depre].mean():.3f}')


  print()

print(f'Desvio padrao Atividades Fisicas: {np.std([dados[atvdd_fisica]])}')
print(f'Desvio padrao Depressao: {np.std([dados[depre]])}')

```
